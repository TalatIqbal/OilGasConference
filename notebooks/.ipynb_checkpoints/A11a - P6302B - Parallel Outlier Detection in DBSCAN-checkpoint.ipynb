```python
import numpy as np

def gauss_2d(mu, sigma, num_samples):
    x = np.random.normal(mu[0], sigma[0], num_samples)
    y = np.random.normal(mu[1], sigma[1], num_samples)
    return (x, y)
```

```python
from matplotlib import pyplot as plt
import math
from sklearn.datasets.samples_generator import make_blobs

x_sin_mu = [i/10 for i in range(0, 5)]
y_sin_mu = [(math.sin(i)+1)*5 for i in x_sin_mu]

print(len(x_sin_mu), len(y_sin_mu))

# plt.figure(figsize=(8,6))
# plt.subplot(1,1,1)
# plt.scatter(x_sin_mu, y_sin_mu, color='g')
# plt.show()

# Generate the dataset
import numpy

sigma = 0.2
mu_x = x_sin_mu
mu_y = y_sin_mu

x, y = gauss_2d([mu_x[0], mu_y[0]], [sigma, sigma], 10)
for i in range(1, len(mu_x)):
    mu = [mu_x[i], mu_y[i]]
    x_samp, y_samp = gauss_2d(mu, [sigma, sigma], 10)
    x,y = np.hstack((x, x_samp)), np.hstack((y, y_samp))

# Add three randomly generated 
x_samp, y_samp = gauss_2d([8, 0.5], [sigma, sigma], 20)
x, y = np.hstack((x, x_samp)), np.hstack((y, y_samp))

# Add random noise
x_samp, y_samp = gauss_2d([5, 5], [3,3], 10)
x, y = np.hstack((x, x_samp)), np.hstack((y, y_samp))


# Add three randomly generated 
x_samp, y_samp = gauss_2d([1, 1], [sigma, sigma], 20)
x, y = np.hstack((x, x_samp)), np.hstack((y, y_samp))

# Add three randomly generated 
x_samp, y_samp = gauss_2d([5, 9], [sigma, sigma], 20)
x, y = np.hstack((x, x_samp)), np.hstack((y, y_samp))

# Add random noise
x_samp, y_samp = gauss_2d([5, 5], [3,3], 10)
x, y = np.hstack((x, x_samp)), np.hstack((y, y_samp))

idx = [i for i in range(0, len(x))]

fig, ax = plt.subplots()
ax.scatter(x,y,color='g')


for i, idx_val in enumerate(idx):
    ax.annotate(idx_val, (x[i], y[i]))
    

plt.show()
```

```python
def find_elem_count(nlist, elem):
    count = 0
    for i in nlist:
        if i == elem:
            count += 1
    return count
```

```python
import scipy.spatial.distance
import pickle as pkl


def dbscan_outliers_sub(X, X_complete, eps, mpts, max_rows_per_iteration, labels_file):

    '''
    Returns an array of labels (outliers are tagged as 0 and others are tagged as 1)
    Note: This does not return the clusters. Everything that is a 1 belongs to 
            some cluster or the other
    @param X - input dataset
    @param eps - epsilon value
    @param mpts - minimum points to be found within the dataset
    @max_rows_per_iteration - dividing the dataset so that the distance metric is 
        computed only among the subset dataset
    '''
    
    n_processed_rows = 0
    start_idx = 0
    total_rows = len(X)
    continue_run = True
        
    labels = list(map(int, np.zeros(len(X_complete)))) # Initialize labels to zeros    
    labels_True = list(map(int, np.zeros(len(X_complete)))) # Initialize labels to zeros
    
    overall_row_idx = 0
    
    while continue_run:
        
        end_idx = start_idx + max_rows_per_iteration
        
        if end_idx >= total_rows:
            end_idx = total_rows
            continue_run = False
        
        tic = time.time()
        X_sub = X[start_idx:end_idx]
        
        print('Distance metric: start=', start_idx, ' end=', end_idx, end ='\t')
        dist = scipy.spatial.distance.cdist(X_sub, X_complete, metric='euclidean', p=2, V=None, VI=None, w=None)
        print('COMPUTED -> Size of dist ', dist.shape, 'in ', str(time.time()-tic), ' secs')
        

        print('Computing neighbours', end='\t')
        for i in range(len(dist)):

            data_instance_dist = enumerate(list(dist[i,:]))
            labels_True = labels.copy()
            local_neighbours = 0
            
            # Iterate through each of the neighbor distance for a given data instance            
            for idx,val in data_instance_dist:
                #if overall_row_idx == 114:                    
                    #print(idx, val, end='\n')
                if val <= eps:
                    local_neighbours += 1
                    labels_True[idx] = labels_True[idx] | 1
                
            # If the number of neighbours found within eps is greater than eps,
            # use labels (altered ones) as real labels and assign that as labels_false 
            # for next iteration
            # Else use labels_False as labels for next iteration
            # At the start of each iteration of data instance, both labels and labels_False should be the same
            if local_neighbours >= mpts:
                labels = labels_True.copy()
            
            overall_row_idx += 1
            
        start_idx = end_idx
        
        print('Outlier Count', find_elem_count(labels, 0), '\tTotal time: ', str(time.time()-tic), ' secs', end='\n')
        
        print('-------------------------------------')
        
    print('Length of the labels:', len(labels))
        
    '''
    str_labels = ','.join(map(str,labels))    
    with open(labels_file, 'w') as f:
        f.write(str_labels)       
    '''
    
    with open(labels_file, 'wb') as f:
        pkl.dump(file=f, obj=labels)
        
    return labels
```

```python
start = 0
end = 50
labels_file = LABELS_DIR + str(start) + '_' + str(end) + '.pkl'
labels = dbscan_outliers_sub(X[start:end], X, 0.5, 3, 5, labels_file)
```

```python
pid = 1551135119

import os
import sys
import pickle as pkl
import numpy as np


labels = ()
pump = 'P6302B'
LABELS_DIR = '../data/results/'+pump+'/labels/'+str(pid)+'/'

for labels_file in os.listdir(LABELS_DIR):
    labels_file = os.path.join(LABELS_DIR, labels_file)
    print(labels_file)

    if labels_file.endswith('.pkl'):
        with open(labels_file, 'rb') as fp:
            sub_labels = pkl.load(fp)

            if len(labels) == 0:
                labels = list(map(int, np.zeros(len(sub_labels))))

            for idx,val in enumerate(labels):
                labels[idx] = labels[idx] | sub_labels[idx]

print(labels)

labels_file = LABELS_DIR + 'complete_labels.pkl'
with open(labels_file, 'wb') as fp:
    pkl.dump(labels, fp)
```
