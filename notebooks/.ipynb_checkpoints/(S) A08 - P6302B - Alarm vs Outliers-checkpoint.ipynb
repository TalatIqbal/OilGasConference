### Plotting comparitive graphs

```python
def plot_time_graphs(df_features, feature_set, y_axis_col):

    if len(feature_set) == 0:
        feature_set = df_features.keys()

    fmt = '%Y-%m-%d %H:%M:%S'
    base_date = datetime.strptime('2014-01-01 00:00:01', fmt)
    s = datetime.strptime('2014-01-01 00:00:01', fmt)
    e = datetime.strptime('2019-01-01 00:00:01', fmt)

    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
    xlabel = 'Time'
    ylabel = 'Value'

    if len(feature_set) == 0:
        feature_set = list(df_features_imp.keys())

    total_plots = len(feature_set)
    cols_plot = 3
    rows_plot = math.ceil(total_plots/cols_plot)

    plt.rcParams['figure.figsize'] = [cols_plot * 10, rows_plot * 7] # Size of the plots        
    plt.figure()

    idx = 0
    for feature in feature_set:
        if feature in df_features.keys():
            df = df_features[feature]

            plt.subplot(rows_plot, cols_plot, idx+1)
            # plt.xlabel(xlabel)
            plt.ylabel(ylabel)
            plt.xticks(rotation=45)
            plt.title('Feature - ' + feature)

            x = df[time_col]
            y = df[y_axis_col]
            plt.plot(x, y, linewidth=1.0, color=colors[idx % len(colors)], alpha=0.5, label='per '+time_granularity+' value')

            plt.legend(loc='upper left')

            idx += 1
```
