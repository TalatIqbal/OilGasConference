### Functions

```python
def get_minutes_after(current_date, base_date):
    
    base_ts = time.mktime(base_date.timetuple()) # Converting to Unix timestamp
    current_ts = time.mktime(current_date.timetuple())
    time_diff_min = round((current_ts - base_ts) / 60.0)
    
    return time_diff_min
```

```python
def get_time_from_minutes(time_in_mins, base_date):    
    new_date = base_date + timedelta(minutes = time_in_mins)        
    return new_date
```

### Generate Master Dataframe

```python
fmt = '%Y-%m-%d %H:%M:%S'
start_date = datetime.strptime('2017-01-01 00:00:01', fmt)
stop_date = datetime.strptime('2019-01-01 00:00:01', fmt)

max_mins = get_minutes_after(stop_date, start_date)
print('Max mins = ', max_mins)

### Generate a dataframe (granularity = per minute)

df_master = pd.DataFrame(columns=['time_in_mins'])

# The dataframe will contain one row for each minute
df_master['time_in_mins'] = [i for i in range(1, max_mins+1)]
```

## Plotting variances

```python
data_dir = INT_DATA_DIR + 'common/'
df_current_tags = {}

pkl_files = [f for f in listdir(data_dir) if (isfile(join(data_dir, f))) and (f.endswith('.pkl'))]
print('Number of files in directory', len(pkl_files))

for pkl_file in pkl_files:
    tag , _ = os.path.splitext(pkl_file)
    pkl_file = data_dir + pkl_file    
    with open(pkl_file, 'rb') as f:
        df = pkl.load(f)
        df_current_tags[tag]= df
print('Number of sub-dataframes in df_current_tags = ', len(df_current_tags))
```

```python
df_stats = pd.DataFrame(columns=['file_base', 'total_count', 'missing_val_count', 'max_val', 'min_val', 'variance', 'std', 'mean_val', 'median_val', 'kurt', 'skew'])
idx = 0

for k, df in df_current_tags.items():    
    df_stats.loc[idx] = [k, len(df), len(df)-df['val'].count(), df['val'].max(), df['val'].min(), df['val'].var(), df['val'].std(), df['val'].mean(), df['val'].median(), df['val'].kurt(), df['val'].skew()]
    idx += 1
```

```python
df_stats
```

```python
df_variances = df_stats[['file_base', 'variance']]
df_variances = df_variances.loc[df_variances['variance'] > 0.05]
df_variances = df_variances.sort_values(['variance'], ascending=False)
x = df_variances['file_base']
y = df_variances['variance']
x_pos = np.arange(len(x))

plt.bar(x_pos, y, align='center', alpha=0.5)
plt.xticks(x_pos, x, rotation=90)
plt.xlabel('Tags')
plt.ylabel('Variance')
plt.title('Variances among different parameters')
 
plt.show()
```

## Plotting comparative graphs - with Values

```python
fmt = '%Y-%m-%d %H:%M:%S'
base_date = datetime.strptime('2017-01-01 00:00:01', fmt)
start_date = datetime.strptime('2017-06-16 00:00:01', fmt)
stop_date = datetime.strptime('2018-10-01 00:00:01', fmt)

max_mins = get_minutes_after(stop_date, base_date)
min_mins = get_minutes_after(start_date, base_date)

df_master.drop(columns=list(set(df_master.columns.values)-set(['time_in_mins'])), axis=1, inplace=True)
df_master_timetrunc = df_master.loc[(df_master['time_in_mins'] < max_mins) & (df_master['time_in_mins'] > min_mins)]

common_dir = INT_DATA_DIR + 'common/'
tags_to_use = ['XI22F25X.PNT', 'XI22F25Y.PNT', 'XI22F26X.PNT','XI22F26Y.PNT', 'XI22F29X.PNT', 'XI22F29Y.PNT', 'XI22F30X.PNT', 'XI22F30Y.PNT']
df_current_tags = {}

# Read the data from pickle files
if len(tags_to_use) > 0:
    for tag in tags_to_use:
        pkl_file = common_dir + tag + '.pkl'
        with open(pkl_file, 'rb') as f:
            df = pkl.load(f)
            df_current_tags[tag]= df
else:
    pkl_files = [f for f in listdir(data_dir) if (isfile(join(data_dir, f))) and (f.endswith('.pkl'))]
    print('Number of files in directory', len(pkl_files))

    for pkl_file in pkl_files:
        tag,_ = os.path.splitext(pkl_file)
        pkl_file = data_dir + pkl_file    
        with open(pkl_file, 'rb') as f:
            df = pkl.load(f)
            df_current_tags[tag]= df


colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
idx = 0

for tag, df_sel in df_current_tags.items():
    df_sel = df_sel[df_sel['val'].notnull()]
    df_sel = df_sel.drop_duplicates(subset='time_in_mins', keep="last")

    df_sel = pd.merge(df_master_timetrunc, df_sel[['read_time', 'time_in_mins', 'scaled_val']], how='left', left_on=['time_in_mins'], right_on=['time_in_mins'])
    df_sel.rename(columns={'scaled_val':tag}, inplace=True)
    df_sel = df_sel.ffill()
    df_sel = df_sel.fillna(0.0)
    
    df_sel['read_time'] = df_sel['time_in_mins'].apply(lambda x: get_time_from_minutes(x, base_date))
        
    df_sel['mov_avg_day'] = df_sel[tag].rolling(1440).mean()
    df_sel['mov_avg_semimonth'] = df_sel[tag].rolling(21600).mean()
    df_sel['mov_avg_month'] = df_sel[tag].rolling(43200).mean()    
        
    plt.xlabel('Time')
    plt.ylabel('scaled value]')
    plt.title('Comparison of vibration features')
    
    x = df_sel['read_time']
    y = df_sel[tag]    
    plt.plot(x, y, linewidth=1.0, color=colors[idx % len(colors)], alpha=0.2, label=tag)
        
    plt.legend(loc='lower right')
        
    idx += 1
```
