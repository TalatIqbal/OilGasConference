```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime as dt
import sys
import time
```

Functions

```python
def days_after(current, since='2016-01-01'):    
            
    date_parts = since.split('-')
    if len(date_parts) == 3:
        since_date = datetime.datetime(int(date_parts[0]), int(date_parts[1]), int(date_parts[2].split()[0]))
        
    date_parts = current.split('-')
    if len(date_parts) == 3:
        current_date = datetime.datetime(int(date_parts[0]), int(date_parts[1]), int(date_parts[2].split()[0]))
        delta = current_date - since_date
        return delta.days    
    else:
        return np.nan    
```

```python
def get_time(dt_str):
    dt_str = dt_str.strip()
    dtobj = dt.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    return dtobj
```

```python
def strip(ip_str):
    return ip_str.strip()
```

Processing Chat Messages

```python
# Read the file as a dataframe
df_chats = pd.read_csv(RAW_PATH+CHAT_FILE, 
                     delimiter='|', 
                     header=None,
                     names=['time', 'sender', 'receiver', 'chat', 'other'])

df_chats = df_chats.rename(columns={'time':'chat_time'})

# Concatenate messages and others (if not null)
df_chats['message'] = np.where(df_chats['other'].isna(), df_chats['chat'], df_chats['chat'] + df_chats['other'])

# Drop the columns
df_chats = df_chats.drop(columns=['chat', 'other'])

# Create a chat_day column that gives the number of days since 2016-01-01
df_chats['chat_day'] = df_chats['chat_time'].apply(lambda x: days_after(x))

# Drop the rows that do not have a valid date - removing nans
df_chats = df_chats.dropna(subset=['chat_day'])

# Convert the column 'time' from string to datetime
df_chats['chat_time'] = pd.to_datetime(df_chats['chat_time'])

df_chats['sender'] = df_chats['sender'].apply(lambda x: strip(x))
df_chats['receiver'] = df_chats['receiver'].apply(lambda x: strip(x))
df_chats['message'] = df_chats['message'].apply(lambda x: strip(x))
```

```python
df_chats.head()
```

Processing Orders

```python
df_orders = pd.read_csv(ORDERS_PATH+ORDERS_FILE, delimiter=',')

df_orders = df_orders[['order_type', 'order_id', 'post_date', 'price', 'quantity', 'market_name',
                      'trader_login', 'broker_login']]

df_orders = df_orders.rename(columns={'post_date':'order_time'})

# Create a order_day column that gives the number of days since 2016-01-01
df_orders['order_day'] = df_orders['order_time'].apply(lambda x: days_after(x))

# Drop the rows that do not have a valid date - removing nans
df_orders = df_orders.dropna(subset=['order_day'])

# Convert the column 'time' from string to datetime
df_orders['order_time'] = pd.to_datetime(df_orders['order_time'])

df_orders['trader_login'] = df_orders['trader_login'].apply(lambda x: strip(x))
df_orders['broker_login'] = df_orders['broker_login'].apply(lambda x: strip(x))

df_orders['prev_orders'] = ''
```

```python
df_orders['prev_order_id'] = df_orders[['','']]
```

Getting a combined dataframe to eliminate invalid orders

```python
df_combined_1 = pd.merge(df_chats, df_orders, 
                       left_on=['sender', 'receiver', 'chat_day'], 
                       right_on=['trader_login', 'broker_login', 'order_day'], 
                       how='inner')

df_combined_2 = pd.merge(df_chats, df_orders, 
                       left_on=['receiver', 'sender', 'chat_day'], 
                       right_on=['trader_login', 'broker_login', 'order_day'], 
                       how='inner')

df_combined = df_combined_1.append(df_combined_2)

# Dont forget to run this statement
df_combined = df_combined[df_combined.chat_time < df_combined.order_time]

lst = [df_combined_1, df_combined_2]
del df_combined_1, df_combined_2
del lst
```

```python
# Pivoting the prev orders values as columns and then assigning multiple labels to each order
df_multilabel = df_multilabel[0:0]

for idx, row in df_market_chat.iterrows():
    
    # print('Idx: ', idx)

    mkt_set = set(row['prev_orders'].split('|'))
    msg = row['message']
    
    empty_series = pd.Series({'message':msg})
    empty_series.name = idx
    df_multilabel = df_multilabel.append(empty_series)

    for mkt in mkt_set:
        mkt = mkt.strip().lower()
        df_multilabel.at[idx, mkt] = 1
        
    # print(df_multilabel[['message', 'cal', 'clk-edm', 'clk-hard']])
        
```

# Alignment for RNN

```python
df_chats.head()
```

```python
# valid orders are the list of all orders that have a chat corresponding to them
# They are mapped using the chats between a broker and a trader pair on a particular day

# Join orders and combined
df_valid_orders = pd.merge(df_orders, df_combined, left_on=['order_id'], right_on=['order_id'], how='inner', suffixes=['','_1'])

# Use only the columns pertaining to order table
df_valid_orders = df_valid_orders[['order_id', 'order_time', 'market_name', 'trader_login', 'broker_login', 'order_day']]

# Drop the duplicates
df_valid_orders.drop_duplicates(inplace=True)
```

```python
df_chats_orders = df_combined[['chat_time', 'order_time', 'message', 'order_id', 'market_name', 'trader_login', 'broker_login', 'order_day']]
```

```python
df_chats_orders.head()
```

```python
# Group the rnn table based on the trader,broker, day combination
df_grouped_chats_orders = df_chats_orders.groupby(['trader_login','broker_login','order_day'])
```

```python
import time
tic = time.time()
index = 0
df_rnn_aligned = pd.DataFrame(columns=['message', 'market', 'order_id', 'trader', 'broker'])

for g, vals in df_grouped_chats_orders:
    
    trader, broker, day = g
    trader = trader.strip()
    broker = broker.strip()
    
    # df_chat_order = df_chats_orders.loc[(df_chats_orders['trader_login'] == trader) & (df_chats_orders['broker_login'] == broker) &  (df_chats_orders['order_day'] == day)]
    
    df_chat = df_chats.loc[(df_chats['sender'] == trader) & (df_chats['receiver'] == broker) & (df_chats['order_day'] == day)].append(
        df_chats.loc[(df_chats['sender'] == broker) & (df_chats['receiver'] == trader) & (df_chats['order_day'] == day)])
    df_chat.sort_values(by=['chat_time'], inplace=True)
    df_chat.reset_index(inplace=True)
    
    df_order = df_valid_orders.loc[(df_valid_orders['trader_login'] == trader) & (df_valid_orders['broker_login'] == broker) & (df_valid_orders['order_day'] == day)]    
        
    print(df_chat[['chat_time', 'message']])
    print(df_order[['order_id', 'order_time', 'market_name']])
    
    # For each order, join all the messages
    for idx, row in df_order.iterrows():
        order_id = row['order_id']
        order_time = row['order_time']
        order_market = row['market_name']
        trader = row['trader_login']
        broker = row['broker_login']
        
        df_complete_chat = df_chat.loc[(df_chat['chat_time']< order_time)]['message'].values
        df_chat = df_chat.loc[(df_chat['chat_time'] > order_time)]
        complete_chat = ' '.join(df_complete_chat)
        
        df_rnn_aligned.at[index, 'message'] = complete_chat
        df_rnn_aligned.at[index, 'market'] = order_market
        df_rnn_aligned.at[index, 'order_id'] = order_id
        df_rnn_aligned.at[index, 'trader'] = trader
        df_rnn_aligned.at[index, 'broker'] = broker
        
        index += 1
    
        
    if (index > 5):
        break;
        
    print(df_rnn_aligned)
    
print('Total seconds taken to run ', (time.time()-tic), 'secs')
```
